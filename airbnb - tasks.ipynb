```python
# Write a custom aggregation function to calculate the mean and standard deviation of the "Price" column



def mean_std_agg(ser):
    return {
        'mean': ser.mean(),
        'std': ser.std(),
        'var': ser.var()
    }

df['Price'].agg(mean_std_agg)

```

```python
# Group the data by "Room_Type" and apply a custom function to calculate the range (max-min) of "cleanliness_rating" for each group.

def range1(ran):
    return ran.max() - ran.min()


# Group by 'Room_Type' and apply the custom function to 'cleanliness_rating'
df.groupby('Room_Type')['cleanliness_rating'].apply(range1)

```
